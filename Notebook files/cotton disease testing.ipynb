```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.models import load_model
```

```python
test_path=r'F:\cotton disease pro\test'
```

```python
##diseased leaf
img=image.load_img(r'F:\cotton disease pro\test\diseased cotton leaf\dis_leaf (193)_iaip.jpg',target_size=(180,180,3))

image_tensor=image.img_to_array(img)

image_tensor.shape

img_ten=np.expand_dims(image_tensor,axis=0)

img_ten.shape

img_ten/=255

plt.imshow(img_ten[0])
plt.show()
```

```python
model=load_model('F:\cotton disease pro\cotton_disease_prediction.h5')
```

```python
model.predict(img_ten)
```

```python
prediction=np.argmax(model.predict(img_ten),axis=1)
```

```python
prediction
```
